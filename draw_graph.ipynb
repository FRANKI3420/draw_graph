## 頂点ラベル付きグラフの描画とファイル出力

```python
import networkx as nx
import matplotlib.pyplot as plt

from matplotlib.backends.backend_pdf import PdfPages


pdf = PdfPages("graph.pdf")

# ファイルを読み込む
with open("query.txt", "r") as f:
    data = f.readlines()
    
data_len = len(data)

now = 0


while True:
# 頂点数を取得   
    now +=1
    labels = None
    n = int(data[now])
    
    # 頂点IDを取得
    nodes = [int(x) for x in data[now+1:now+1+n]]

    # グラフを初期化
    G = nx.Graph()
    
    id = 0


    # 頂点をグラフに追加
    for node in nodes:
        # print(node)
        # G.nodes[id] = node
        G.add_node(id)
        if id==0:
            labels = {id: node}
        else:
            labels[id] = node

        id+=1



    # 辺の数を取得
    m = int(data[now+n+1])
    
    # 辺をグラフに追加
    for line in data[now+n+2:now+n+2+m]:
        u, v = map(int, line.split())
        G.add_edge(u, v)
        
    now+=n+2+m
    
    fig = plt.figure()

  # force-directed layoutアルゴリズムを使用してノードの位置を計算
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_labels(G, pos, labels,font_color='white')
    
    # plt.show()
    
    plt.close()   # Figureオブジェクトを閉じる
    pdf.savefig(fig)
        
    if now>=data_len:
        break
pdf.close()
```
